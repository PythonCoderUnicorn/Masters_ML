## Filter column by category 
- ages of all non-smokers

```python
non_smoker_age = df[df.Smoke == 0]['Age'].to_numpy()
avg_age_non_smoker = non_smoker_age.mean()
print('{:.2f} is average age for non smokers'.format(avg_age_non_smoker)
```

```python
# find non-smokers who answered 3 on survey
non_Smoker_satisfied_age = df.loc[ (df.Smoke ==0) & (df.Survey > 3),'Age'].to_numpy()

# find non-smokers who survey != 3
criteria = (df.Smoke == 0) & (df.Survey > 3)
non_Smoker_NotSatisfied_age = df.loc[~criteria, 'Age'].to_numpy()
```

## Rename Column names

```python
df.rename(columns= {'Name':'Patient'}, inplace=True)
df.columns
```

## Add value to Age column for privacy 
- ``df.Age.apply(add2_func)``
- or quick way: ``df.Age = df.Age.apply(lambda x: x + 2)``

```python
# adding 2 to Age column function
def add2_func(x):
    return x + 2

df.Age = df.Age.apply(add2_func)
df.Age.head()
```

## Change Nominal variables to Ordinal

```python
# check what nominal vars are in Group
df.Group.unique()
```

Control group will be 0 and Active will be 1

```python
# map() is like a for loop and deletes unspecified data
# replace() keeps unspecified data
df.Group = df.Group.replace({'Control':0, 'Active':1})
df.Group.head()
```

## Change Categorical variable to Numerical
- *CholesterolBefore* column has 3 categories: [low, intermediate, high]

```python
print('Chol min:',df.CholesterolBefore.min())
print('Chol max:',df.CholesterolBefore.max())

# use bins= 3 for equal sized bins
df['CholesterolBeforeLevel'] = pd.cut(df.CholesterolBefore, bins=3, 
                                      labels=['low', 'intermediate', 'high'])

# df[['CholesterolBefore', 'CholesterolBeforeLevel']].head(10).to_numpy() 

# create three bins with intervals 
# low = [0,5), intermediate = [5,10), and high = [10,20)
df.CholesterolBeforeLevel = pd.cut(df.CholesterolBefore, 
                                   bins=[0,5,10,20], 
                                   right=False, 
                                   labels=['low', 'intermediate', 'high'])
```

## Split Column string values

```python
new_df = df.Patient.str.split(' ', expand=True)
df['FirstName'] = new_df[0]
df['LastName'] = new_df[1]

# join the split names into 1 with last name first
df['Patient'] = df.LastName + ', '+ df.FirstName
df.head()
```
